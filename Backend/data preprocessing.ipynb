```python
import os
from glob import glob
import pandas as pd
from functools import reduce
from xml.etree import ElementTree as et
```

```python
#load all xml files and store in a list
xml_list=glob('C:/Hrushi/tri_nit_24/Japan/train/annotations/xmls/*.xml')
#data cleaning
xml_list=list(map(lambda x: x.replace('\\','/'),xml_list))
```

```python
xml_list
```

```python
#extract file name,size(height,width),object(name,x_min,x_max,y_min,y_max)
def extract_text(filename):
    tree = et.parse(filename)
    root=tree.getroot()

    #extract filename

    image_name = root.find('filename').text

    #width and height

    width = root.find('size').find('width').text
    height=root.find('size').find('height').text

    objs = root.findall('object')

    parser = []

    for obj in objs:
        name=obj.find('name').text
        bndbox=obj.find('bndbox')
        x_min=bndbox.find('xmin').text
        x_max=bndbox.find('xmax').text
        y_min=bndbox.find('ymin').text
        y_max=bndbox.find('ymax').text
        parser.append([image_name, width, height, name,x_min,x_max,y_min,y_max])
    return parser
```

```python
parser_all=list(map(extract_text,xml_list))
```

```python
parser_all
```

```python
data = reduce(lambda x, y: x+y, parser_all)
data
```

```python
df = pd.DataFrame(data,columns = ['filename','width','height','name','xmin','xmax','ymin','ymax'])
```

```python
#type conversions

col=['width','height','xmin','xmax','ymin','ymax']
df[col] = df[col].astype(int)
df.info()
```

```python
#center x, center y
df['center_x']=((df['xmax']+df['xmin'])/2)/df['width']
df['center_y']=((df['ymax']+df['ymin'])/2)/df['height']

#width and height
df['w']=(df['xmax']-df['xmin'])/df['width']
df['h']=(df['ymax']-df['ymin'])/df['height']
```

```python
df.head(10)
```

```python
#split data into train and test
images=df['filename'].unique()
len(images)
```

```python
image_df=pd.DataFrame(images,columns=['filename'])
img_train=tuple(image_df.sample(frac=0.8)['filename'])
```

```python
img_test=tuple(image_df.query(f'filename not in {img_train}')['filename'])
```

```python
len(img_train),len(img_test)
```

```python
train_df = df.query(f'filename in {img_train}')
test_df = df.query(f'filename in {img_test}')
```

```python
#label Encoding
def label_encoding(x):
    labels = { 'D00':0, 'D10':1, 'D20':2, 'D40':3, 'D43':4, 'D44':5, 'D50':6}
    return labels[x]
```

```python
train_df['id']=train_df['name'].apply(label_encoding)
test_df['id']=test_df['name'].apply(label_encoding)
```

```python
test_df.head(10)
```

```python
import os
from shutil import move
```

```python
train_folder='C:/Hrushi/tri_nit_24/Japan/train/images/train'
test_folder='C:/Hrushi/tri_nit_24/Japan/train/images/test'

os.mkdir(train_folder)
os.mkdir(test_folder)
```

```python
cols=['filename','id','center_x','center_y','w','h']
groupby_obj_train=train_df[cols].groupby('filename')
groupby_obj_test=test_df[cols].groupby('filename')
```

```python

#save each image in train/test folder and respective labels in it

def save_data(filename,folder_path,group_obj):
    #move image
    src = os.path.join('C:/Hrushi/tri_nit_24/Japan/train/images',filename)
    dst = os.path.join(folder_path,filename)
    move(src,dst) # move images to dst folder
    
    #save label info
    text_filename = os.path.join(folder_path, os.path.splitext(filename)[0]+ '.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename,sep=' ', index=False,header=False)
```

```python
filename_series=pd.Series(groupby_obj_train.groups.keys())
```

```python
filename_series.apply(save_data,args=(train_folder,groupby_obj_train))
```

```python
filename_series_test=pd.Series(groupby_obj_test.groups.keys())
filename_series_test.apply(save_data,args=(test_folder,groupby_obj_test))
```
